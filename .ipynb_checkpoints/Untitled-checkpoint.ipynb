```python
def strip_string(s):
    s = s.strip()
    s = s.replace('\n', '')
    for i in range(5):
        s = s.replace('  ', ' ')
    return s
```

# Table 7 Content 

```python
data = formatted_data.findAll("section", {"class": "certificates"})[0].findAll("ul", {"class": "timeline"})
n = 1

for i in range(len(data)):
    tid = 'TB70' + str(n)
    
    certificate_name = data[i].h3.text
    issuer = data[i].findAll('h3')[1].text
    location = data[i].p.text
    certificate_link = data[i].a.get('href')
    start_date = data[i].findAll('span')[0].text
    end_date = data[i].findAll('span')[1].text
    duration = data[i].findAll('span')[2].text
    details = data[i].findAll('p')[1].text
    
    
    print('TID -> ', tid)
    print('certificate_name -> ', strip_string(certificate_name))
    print('Issuer -> ', issuer)
    print('Location -> ', strip_string(location))
    print('certificate_link -> ', strip_string(certificate_link))
    print('Start Date -> ', strip_string(start_date))
    print('End Date -> ', strip_string(end_date))
    print('Duration -> ', strip_string(duration))
    print('Details -> ', strip_string(details))
    
    print('-' * 100)
    n += 1
    
    
    query = f"""INSERT INTO CERTIFICATES (TID, CERTIFICATE_NAME, ISSUER, LOCATION, CERTIFICATE_LINK, START_DATE, END_DATE, DURATION, DETAILS) VALUES ("{tid}","{strip_string(certificate_name)}", "{issuer}", "{strip_string(location)}", "{strip_string(certificate_link)}", "{strip_string(start_date)}", "{strip_string(end_date)}", "{strip_string(duration)}", "{strip_string(details)}")"""
    print(query)
    mycursor.execute(query)
    mycursor.commit()
```
